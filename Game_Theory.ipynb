**Scenario:** Two companies are competing in a market by deciding how much to spend on advertising. Write a Python program that simulates this situation using game theory and calculates the Nash equilibrium for various spending strategies.

```python
import numpy as np
from nashpy import Game  # Library for solving game theory problems
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Define the payoff matrices
# Example: Advertising budgets in millions [0, 1, 2]
strategies = [0, 1, 2]

# Payoff matrices for Company A and Company B
payoff_A = np.array([
    [3, 2, 1],  # A’s payoffs when B spends [0, 1, 2]
    [4, 3, 2],
    [5, 4, 3]
])

payoff_B = np.array([
    [3, 4, 5],  # B’s payoffs when A spends [0, 1, 2]
    [2, 3, 4],
    [1, 2, 3]
])
```

```python
# Step 2: Create a game object
game = Game(payoff_A, payoff_B)
```

```python

# Step 3: Find Nash equilibria
equilibria = list(game.support_enumeration())
```

```python
# Display results
if equilibria:
    print("Nash Equilibria Found:")
    for i, eq in enumerate(equilibria):
        print(f"\nEquilibrium {i+1}:")
        print(f"Company A's Strategy: {eq[0]}")
        print(f"Company B's Strategy: {eq[1]}")
else:
    print("No Nash equilibrium found.")
```

```python
# Step 4: Simulate and display payoffs
for i, eq in enumerate(equilibria):
    a_strategy, b_strategy = eq
    payoff_A_result = np.dot(a_strategy, np.dot(payoff_A, b_strategy))
    payoff_B_result = np.dot(a_strategy, np.dot(payoff_B, b_strategy))
    print(f"\nPayoffs for Equilibrium {i+1}:")
    print(f"Company A's Payoff: {payoff_A_result}")
    print(f"Company B's Payoff: {payoff_B_result}")
```

```python
# Step 5: Visualize payoff matrices
plt.figure(figsize=(12, 6))
```

```python
# Heatmap for Company A
plt.subplot(1, 2, 1)
sns.heatmap(payoff_A, annot=True, cmap="coolwarm",
            cbar=True, xticklabels=strategies,
            yticklabels=strategies)
plt.title("Company A's Payoff Matrix")
plt.xlabel("Company B's Strategies")
plt.ylabel("Company A's Strategies")
```

```python
# Heatmap for Company B
plt.subplot(1, 2, 2)
sns.heatmap(payoff_B, annot=True, cmap="coolwarm",
            cbar=True, xticklabels=strategies,
            yticklabels=strategies)
plt.title("Company B's Payoff Matrix")
plt.xlabel("Company B's Strategies")
plt.ylabel("Company A's Strategies")

plt.tight_layout()
plt.show()
```
